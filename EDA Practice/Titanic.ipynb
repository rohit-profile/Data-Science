```python
plt.figure(figsize=(12,5))

# Histogram
plt.subplot(1,2,1)
sns.histplot(df['Age'], bins=30, kde=True, color='skyblue')
plt.title('Age Distribution')

# Boxplot
plt.subplot(1,2,2)
sns.boxplot(y=df['Age'], color='lightgreen')
plt.title('Age Spread / Outliers')

plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Fare'], bins=30, kde=True, color='orange')
plt.title('Fare Distribution')
plt.show()
```

```python
sns.countplot(x='Survived', data=df, palette='Set2')
plt.title('Survival Count (0=No, 1=Yes)')
plt.show()
```

```python
sns.countplot(x='Sex', hue='Survived', data=df, palette='Set1')
plt.title('Survival by Gender')
plt.show()
```

```python
sns.countplot(x='Pclass', data=df, palette='Set3')
plt.title('Passengers by Class')
plt.show()
```

```python
df['Embarked'].value_counts().plot.pie(autopct='%1.1f%%', colors=['skyblue','orange','green'])
plt.title('Passengers by Embarked Port')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```
